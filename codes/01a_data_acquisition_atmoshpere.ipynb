```python
import requests
from pathlib import Path
```

```python
energy_type = "solar"
feature = "rsds"
# output directory
outdir = Path(f"../data/raw/BARRA_R2/{energy_type}/{feature}")
outdir.mkdir(exist_ok=True)
```

```python
def download_from_nci(energy_type, feature):
    # define output directory
    dir_output = Path(f"../data/raw/BARRA_R2/{energy_type}/{feature}")
    dir_output.mkdir(exist_ok=True)

    url_base = ("https://thredds.nci.org.au/thredds/fileServer/"
        "ob53/output/reanalysis/AUS-11/BOM/ERA5/historical/hres/"
        "BARRA-R2/v1/1hr/{feature}/latest/"
        "{feature}_AUS-11_ERA5_historical_hres_BOM_BARRA-R2_v1_1hr_{ym}-{ym}.nc")
    # define year range
    for year in range(2014,2024):
        urls = [url_base.format(feature=feature, ym=f"{year}{month:02d}") for month in range(1, 13)]
        # download data
        for url in urls:
            filename = url.split("/")[-1]
            path_output = dir_output / filename
            print(f"Downloading {url} -> {path_output}")
            r = requests.get(url, stream=True)
            with open(path_output, "wb") as f:
                for chunk in r.iter_content(chunk_size=8192):
                    if chunk:
                        f.write(chunk)

        print(f"{year} {feature} downloaded")
```

```python
download_from_nci('solar', 'rsds')
```

```python
download_from_nci("solar","rsdscs")
```
