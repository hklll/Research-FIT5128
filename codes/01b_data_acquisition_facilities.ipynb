```python
directory_facilities = "../data/raw/facilities/"
```

```python
# define a function to convert response into dataframe
def response_to_df(response):
    records = [f.model_dump() for f in response.data]
    df = pd.DataFrame(records)
    return df
```

```python
# obtain solar facilities information
response_solar = await client._async_get_facilities(
    fueltech_id=[UnitFueltechType.SOLAR_UTILITY]
)
# obtain wind facilities information
response_wind = await client._async_get_facilities(
    fueltech_id=[UnitFueltechType.WIND]
)
```

```python
# convert response to dataframe
df_wind = response_to_df(response_wind)
df_solar = response_to_df(response_solar)
```

```python
# save raw data
df_solar.to_parquet(directory_facilities + "facilities_solar.parquet", index=False)
df_wind.to_parquet(directory_facilities + "facilities_wind.parquet", index=False)
```
