```python
# import libraries
from pathlib import Path
import pandas as pd
import os
```

```python
dir_rsds = Path("../data/processed/rsds")
files_rsds = list(dir_rsds.glob("*"))
dict_files = {file.stem: pd.read_parquet(file) for file in files_rsds}
```

```python
threshold_rsds = 120

dict_rsds = {}

for code, df in dict_files.items():
    df = df.loc[(df['rsds'] > threshold_rsds) & (df['bnds']==0.0)]
    df = df[['time','rsds']]
    dict_rsds[code] = df
```

```python
out_dir = "../data/processed/rsds_filter"
os.makedirs(out_dir, exist_ok=True)

for key, df in dict_rsds.items():
    file_path = os.path.join(out_dir, f"{key}.parquet")
    df.to_parquet(file_path, index=False)
```
